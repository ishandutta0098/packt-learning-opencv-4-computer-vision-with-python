# Reading/Writing an Image File

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
def plot(img):
    if len(img.shape) == 2:
        plt.imshow(img, cmap='gray')
        plt.axis('off') 

    else:
        plt.imshow(img)
        plt.axis('off') 
```

```python
# Black Box

img = np.zeros((3,3), dtype=np.uint8)
plot(img)
```

```python
# Black Box 2D to 3D BGR
img = cv2.cvtColor(img, cv2.COLOR_GRAY2BGR)
plot(img)
```

```python
# Load Image 3 CHannel BGR 
img = cv2.imread('MyPic.png', cv2.IMREAD_COLOR)
plot(img)
```

```python
# Load Grayscale

img = cv2.imread('MyPic.png', cv2.IMREAD_GRAYSCALE)
plot(img)
```

```python
# Either BGR or Grayscale depending on Image
img = cv2.imread('MyPic.png', cv2.IMREAD_ANYCOLOR)
plot(img)
```

```python
# Read all channels including aplha/transparency as 4th channel
img = cv2.imread('MyPic.png', cv2.IMREAD_UNCHANGED)
plot(img)
```

```python
# Grayscale at original bit depth 
img = cv2.imread('MyPic.png', cv2.IMREAD_ANYDEPTH)
plot(img)
```

```python
# BGR Image with original bit depth 
img = cv2.imread('MyPic.png', cv2.IMREAD_ANYDEPTH | cv2.IMREAD_COLOR)
plot(img)
```

```python
# Half Resolution Grayscale
img = cv2.imread('MyPic.png', cv2.IMREAD_REDUCED_GRAYSCALE_2)
plot(img)
```

```python
# Half Resolution Color
img = cv2.imread('MyPic.png', cv2.IMREAD_REDUCED_COLOR_2)
plot(img)
```

```python
# One Quarter Resolution Grayscale
img = cv2.imread('MyPic.png', cv2.IMREAD_REDUCED_GRAYSCALE_4)
plot(img)
```

```python
# One Quarter Resolution Color
img = cv2.imread('MyPic.png', cv2.IMREAD_REDUCED_COLOR_4)
plot(img)
```

```python
# One Eighth Resolution Grayscale
img = cv2.imread('MyPic.png', cv2.IMREAD_REDUCED_GRAYSCALE_8)
plot(img)
```

```python
# One Eighth Resolution Color
img = cv2.imread('MyPic.png', cv2.IMREAD_REDUCED_COLOR_8)
plot(img)
```

```python
img = cv2.imread("MyPic.png")
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
```
